```python
def tinh_giai_thua():
    so_nhap = input("Vui lòng nhập một số nguyên không âm: ")
    n = int(so_nhap)
    
    if n < 0:
        print("Không thể tính giai thừa cho số âm.")
        return
    elif n == 0:
        ket_qua = 1
    else:
        ket_qua = 1
        for i in range(1, n + 1):
            ket_qua = ket_qua * i

    print(str(n) + " giai thừa = " + str(ket_qua))

tinh_giai_thua()
```
