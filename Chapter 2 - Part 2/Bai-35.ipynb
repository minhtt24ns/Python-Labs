```python
    num = input("Vui lòng nhập một chuỗi các số, cách nhau bằng dấu phẩy (ví dụ: 1,2,3,...): ")

    danh_sach_chuoi = [x.strip() for x in num.split(',')]

    so_le = []

    for phan_tu in danh_sach_chuoi:

            so = int(phan_tu)
            
            if so % 2 != 0:
                so_le.append(so)

    ket_qua = ",".join(str(n) for n in so_le)
    
    print("Chuỗi các số lẻ: " + ket_qua)
```
