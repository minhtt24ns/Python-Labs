```python
def tao_danh_sach_va_cat():
    danh_sach_ket_qua = []

    for i in range(1, 21):
        binh_phuong = i ** 2
        
        danh_sach_ket_qua.append(binh_phuong)
    
    nam_phan_tu_dau = danh_sach_ket_qua[0:5]
    
    print("Danh sách 5 phần tử đầu tiên:")
    print(nam_phan_tu_dau)

tao_danh_sach_va_cat()
```
