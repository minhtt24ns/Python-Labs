```python

    danh_sach_so = []

    print("Nhập các số nguyên (nhập 0 để kết thúc và hiển thị kết quả):")

    while True:
        dau_vao = input()
        chuoi_kiem_tra = dau_vao.strip()

        if chuoi_kiem_tra == "0":
            break
            
 
        so = int(chuoi_kiem_tra)
        danh_sach_so.append(so)

    if not danh_sach_so:
        print("Không có số nguyên nào được nhập (ngoại trừ 0).")

    danh_sach_so.sort()

    print("\nCác giá trị đã nhập (theo thứ tự tăng dần):")
    for gia_tri in danh_sach_so:
        print(gia_tri)
```
