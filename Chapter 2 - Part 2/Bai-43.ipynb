```python
def in_chuoi_dai_hon(chuoi1, chuoi2):
    do_dai1 = len(chuoi1)
    do_dai2 = len(chuoi2)

    if do_dai1 > do_dai2:
        print(chuoi1)
    elif do_dai2 > do_dai1:
        print(chuoi2)
    else:
        print(chuoi1)
        print(chuoi2)

chuoi_dau_vao_1 = input("Nhập chuỗi thứ nhất: ")
chuoi_dau_vao_2 = input("Nhập chuỗi thứ hai: ")

print("\n Kết quả so sánh: ")
in_chuoi_dai_hon(chuoi_dau_vao_1, chuoi_dau_vao_2)
```
