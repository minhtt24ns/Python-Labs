```python

    nam = int(input("Vui lòng nhập một năm (ví dụ: 2024): "))

    la_nam_nhuan = False

    if nam % 400 == 0:
        la_nam_nhuan = True
    elif nam % 100 == 0:
        la_nam_nhuan = False
    elif nam % 4 == 0:
        la_nam_nhuan = True
    else:
        la_nam_nhuan = False

    if la_nam_nhuan:
        print("Năm " + str(nam) + " là năm nhuận.")
    else:
        print("Năm " + str(nam) + " không phải là năm nhuận.")
```
