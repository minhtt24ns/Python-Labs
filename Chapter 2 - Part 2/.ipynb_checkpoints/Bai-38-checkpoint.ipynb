```python
    danh_sach_so = []
    
    print("Vui lòng nhập các số nguyên, mỗi số một dòng (nhập một dòng trống để kết thúc):")
    
    while True:
        dau_vao = input()
        chuoi_kiem_tra = dau_vao.strip()
        
        if chuoi_kiem_tra == "":
            break
            
        so = int(chuoi_kiem_tra)
        danh_sach_so.append(so)


    so_am = []
    so_khong = []
    so_duong = []

    for so in danh_sach_so:
        if so < 0:
            so_am.append(so)
        elif so == 0:
            so_khong.append(so)
        else:
            so_duong.append(so)
            
    ket_qua_cuoi = so_am + so_khong + so_duong

    print("\nKết quả sắp xếp:")
    print(", ".join(str(s) for s in ket_qua_cuoi))
```
