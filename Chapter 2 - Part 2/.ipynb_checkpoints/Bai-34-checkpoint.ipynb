```python
    so_thap_phan = int(input("Nhập một số thập phân: "))

    if so_thap_phan == 0:
        print("Số thập phân" + so_thap_phan + "tương đương với số nhị phân: 0")

    nhi_phan = ""
    so_tam = so_thap_phan
    
    while so_tam > 0:
        phan_du = so_tam % 2
        
        nhi_phan = str(phan_du) + nhi_phan
        
        so_tam = so_tam // 2
        
    print("Số thập phân" + str(so_thap_phan) + "tương đương với số nhị phân: "+ nhi_phan)
```
