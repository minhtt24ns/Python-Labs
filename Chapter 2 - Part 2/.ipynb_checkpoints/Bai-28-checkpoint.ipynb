```python
    so_ngay_thang = {
        'tháng một': 31,
        'tháng hai': '28 hoặc 29',
        'tháng ba': 31,
        'tháng tư': 30,
        'tháng năm': 31,
        'tháng sáu': 30,
        'tháng bảy': 31,
        'tháng tám': 31,
        'tháng chín': 30,
        'tháng mười': 31,
        'tháng mười một': 30,
        'tháng mười hai': 31
    }

    ten_thang_nhap = input("Vui lòng nhập tên của một tháng (ví dụ: tháng một, tháng hai, ...): ")

    ten_thang_chuan_hoa = ten_thang_nhap.strip().lower()

    if ten_thang_chuan_hoa in so_ngay_thang:
        so_ngay = so_ngay_thang[ten_thang_chuan_hoa]
        print("Số ngày trong " + str(ten_thang_nhap) + " là " + str(so_ngay))
    else:
        print("Lỗi: Tên tháng không hợp lệ. Vui lòng thử lại.")
```
