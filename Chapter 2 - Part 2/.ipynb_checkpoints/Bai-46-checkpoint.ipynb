```python
def tao_danh_sach_va_cat():
    danh_sach_ket_qua = []

    for i in range(1, 21):
        binh_phuong = i ** 2
        
        danh_sach_ket_qua.append(binh_phuong)
    
    nam_phan_tu_cuoi = danh_sach_ket_qua[-5:]
    
    print("Danh sách 5 phần tử cuối cùng:")
    print(nam_phan_tu_cuoi)

tao_danh_sach_va_cat()
```
