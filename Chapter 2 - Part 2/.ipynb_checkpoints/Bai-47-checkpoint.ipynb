```python
def tao_danh_sach_va_cat():
    danh_sach_ket_qua = []

    for i in range(1, 21):
        binh_phuong = i ** 2
        
        danh_sach_ket_qua.append(binh_phuong)
    
    ds_moi = danh_sach_ket_qua[5:]
    
    print("Danh sách trừ 5 phần tử đầu tiên: ")
    print(ds_moi)

tao_danh_sach_va_cat()
```
