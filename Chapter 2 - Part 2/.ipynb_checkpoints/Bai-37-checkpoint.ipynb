```python

    ds = []
    
    print("Vui lòng nhập các từ (nhập một dòng trống để kết thúc):")
    
    while True:
        st = input()
        
        if st == "":
            break
            

        ds.append(st.strip())

    ket_qua = []
    
    for i in ds:
        if i not in ket_qua:
            ket_qua.append(i)

    print("\nCác từ đã nhập (không trùng lặp, theo thứ tự nhập):")
    for i in ket_qua:
        print(i)

```
