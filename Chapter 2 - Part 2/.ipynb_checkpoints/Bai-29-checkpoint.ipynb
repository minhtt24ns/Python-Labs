```python
    a = float(input("Nhập độ dài cạnh thứ nhất (a): "))
    b = float(input("Nhập độ dài cạnh thứ hai (b): "))
    c = float(input("Nhập độ dài cạnh thứ ba (c): "))

    if a <= 0 or b <= 0 or c <= 0:
        print("Độ dài các cạnh phải là số dương.")

    if a + b <= c or a + c <= b or b + c <= a:
        print("Ba độ dài này không tạo thành một tam giác hợp lệ.")
    else:
        if a == b and b == c:
            print("Đây là một Tam giác đều.")
        elif a == b or a == c or b == c:
            print("Đây là một Tam giác cân.")
        else:
            print("Đây là một Tam giác thường")
```
