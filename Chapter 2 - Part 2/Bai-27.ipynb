```python
n = int(input("Nhập vào số cạnh (từ 3-10): "))

if n==3:
    print("Hình tam giác")
elif n==4:
    print("Hình tứ giác")
elif n==5:
    print("Hình ngũ giác")
elif n==6:
    print("Hình lục giác")
elif n==7:
    print("Hình thất giác")
elif n==8:
    print("Hình bát giác")
elif n==9:
    print("Hình cửu giác")
elif n==10:
    print("Hình thập giác")
else:
    print("Vui lòng nhập số từ 3-10")
```
