```python
import math

a = float(input("Nhap a: "))
b = a
c = a

A = float(input("Nhap goc A: "))
B = float(input("Nhap goc B: "))
C = float(input("Nhap goc C: "))
while (A+B+C!=180) :
    A = float(input("Nhap lai goc A: "))
    B = float(input("Nhap lai goc B: "))
    C = float(input("Nhap lai goc C: "))
S1 = (1/2)*a*b*math.sin(C)
S2 = (1/2)*a*c*math.sin(B)
S3 = (1/2)*c*b*math.sin(A)

print("Dien tich tam giac 1: ",S1)
print("Dien tich tam giac 2: ",S2)
print("Dien tich tam giac 3: ",S3)
```
